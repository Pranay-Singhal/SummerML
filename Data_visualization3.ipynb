```
import pandas as pd
import matplotlib.pyplot as plt 
```

```
df=pd.read_csv("bank.csv")
```

```
df.info()
```

```
df
```

```
# plotting all the possible graphs
plt.scatter(df['CreditScore'],df['RowNumber'])
plt.legend()
```

```
# plotting line graph
plt.plot(df['CustomerId'],df['CreditScore'])
```

```
# such a large data so plotting line graph of top 500 data
plt.plot(df['CustomerId'].head(500),df['CreditScore'].head(500))
```

```
# plotting bar graph b/w RowNumber and estimated salary of top 500 customers
plt.bar(df['RowNumber'].head(500),df['EstimatedSalary'].head(500))
```

```
# plotting pie chart of Geography which will show how much customers belongs to a particular country
exp=(0.1,0.1,0.1)
plt.pie(df['Geography'].value_counts(),labels=['France','Germany','Spain'],explode=exp,shadow=True,autopct='%1.1f%%')
```

```
# Histogram of all numeric data
df.hist()
```
